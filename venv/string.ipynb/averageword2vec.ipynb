```python
stem=PorterStemmer()
```

```python
paragraph=[]
for i in range(len(pf)):
    review=re.sub('[^a-zA-Z]',' ',pf['text'][i])
    review=review.lower()
    review=review.split()
    review=[stem.stem(word) for word in review if word not in stopwords.words("english")]
    review=' '.join(review)
    paragraph.append(review)
```

```python
from gensim.models import Word2Vec
```

```python
max_vocab=500
```

```python
vector=Word2Vec(max_vocab_size=max_vocab,sentences=sent,vector_size=100,min_count=1)
```

```python
import numpy as np
```

```python
def avgword2vec(doc):
    return np.mean([vector.wv[word] for word in doc if word in vector.wv.index_to_key],axis=0)
```

```python
from tqdm import tqdm
```

```python
x=[]
for i in tqdm(range(len(sent))):
    x.append(avgword2vec(sent[i]))
```
