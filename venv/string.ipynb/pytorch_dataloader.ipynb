```python
from sklearn.datasets import make_classification
import torch
```

```python
x,y=make_classification(
    n_samples=20,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_repeated=0,
    n_classes=2,
    random_state=4

)
```

```python
y
```

```python
x=torch.tensor(x,dtype=torch.float32)
y=torch.tensor(y,dtype=torch.float32)
```

```python
x
```

```python
y
```

```python
import torch.nn as nn
```

```python
class mymodel(nn.Module):
    def __init__(self,new_features,hidden_layer):
        super().__init__()
        self.linear1=nn.Linear(new_features,hidden_layer)
        self.relu=nn.ReLU()
        self.linear2=nn.Linear(hidden_layer,1)
        self.sigmoid=nn.Sigmoid()

    def forward(self,features):
        out=self.linear1(features)
        out=self.relu(out)
        out=self.linear2(out)
        out=self.sigmoid(out)
        return out
```

```python
x.shape[1]
```

```python
model=mymodel(x.shape[1],3)
```

```python
from torch.utils.data import DataLoader,Dataset
```

```python
class SimpleNN(Dataset):
    def __init__(self,feature,label):
        self.feature=feature
        self.label=label

    def __len__(self):
        return self.feature.shape[0]
    
    def __getitem__(self, index):
        return self.feature[index],self.label[index]
    
```

```python
x_dataset=SimpleNN(x,y)
```

```python
len(x)
```

```python
x_dataset[2]
```

```python
import os
os.cpu_count()
```

```python
loader=DataLoader(
    dataset=x_dataset,
    batch_size=2,
    shuffle=True)
```

```python
loader
```

```python
import torch.optim as optim
```

```python
learning_rate=0.01
```

```python
epochs=30
```

```python
loss_function=nn.BCELoss()
```

```python
optimizer=optim.Adam(model.parameters(),lr=learning_rate)
```

```python
for epoch in range(epochs):
    for batch_data,batch_label in loader:
        input=batch_data
        output=batch_label
        print(input)
        print(output)
        y_pred=model(input)

        loss=loss_function(y_pred,output.view(-1,1))

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

    print(f" the number of epochs {epoch+1} and loss is {loss}")
```
