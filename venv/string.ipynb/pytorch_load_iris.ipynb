```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
from torch.utils.data import DataLoader,Dataset
import torch.optim as optim
import pandas as pd
import numpy as np
```

```python
data=datasets.load_iris()
```

```python
df=pd.DataFrame(data=data.data,columns=data.feature_names)
df["label"]=data.target
```

```python
x=df.iloc[:,0:4]
```

```python
y=np.array(df['label'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
ss=StandardScaler()
x_train=ss.fit_transform(x_train)
x_test=ss.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
x_train_tensor=torch.tensor(x_train,dtype=torch.float32)
x_test_tensor=torch.tensor(x_test,dtype=torch.float32)
y_train_tensor=torch.tensor(y_train,dtype=torch.float32)
y_test_tensor=torch.tensor(y_test,dtype=torch.float32)
```

```python
y_test_tensor.shape
```

```python
class model(nn.Module):
    def __init__(self, new_features):
        super().__init__()
        self.linear1=nn.Linear(new_features,3)
        self.relu=nn.ReLU()
        self.linear2=nn.Linear(3,1)
    
    def forward(self,features):
        out=self.linear1(features)
        out=self.relu(out)
        out=self.linear2(out)
        return out
```

```python
x_train.shape[1]
```

```python
object=model(x_train.shape[1])
```

```python
class custombatch(Dataset):
    def __init__(self,features,label):
        self.features=features
        self.label=label

    def __len__(self):
        return self.features.shape[0]
    
    def __getitem__(self, index):
        return self.features[index],self.label[index]
```

```python
dataset1=custombatch(x_train_tensor,y_train_tensor)
dataset2=custombatch(x_test_tensor,y_test_tensor)
```

```python
len(dataset1)
```

```python
dataset1[2]
```

```python
train_dataset=DataLoader(dataset1,batch_size=10,shuffle=True)
test_dataset=DataLoader(dataset2,batch_size=10,shuffle=False)
```

```python
learning_rate=0.1
```

```python
epochs=20
```

```python
loss_function=nn.CrossEntropyLoss()
```

```python
optimizer=optim.SGD(object.parameters(),lr=learning_rate)
```

```python
for epoch in range(epochs):
    loss_calculate=0
    for batch_data,batch_label in train_dataset:
        input=batch_data
        label=batch_label

        y_pred=object(input)

        loss=loss_function(y_pred,label.view(-1,1))

        optimizer.zero_grad()

        loss.backward()

        optimizer.step()

        loss_calculate=loss_calculate+loss.item()
        print(loss_calculate)
    print(f"{epoch+1} : {loss_calculate/len(x_train_tensor)}")
```
